According to Csikszentmihalyi, flow occurs when a person's abilities are fully involved in overcoming a challenge that is just about manageable.

- Yellow Zone (Flow Channel): Where the skill and challenge are balanced. This is the zone where flow happens, and it's often depicted as a channel or curve where the relationship between challenge and skill is matched.
